```python
# str and repr
class person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def __str__(self):
        return f"Person(name={self.name}, age={self.age})"
    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"
p = person("Arslan",22)
print(p) # calls __str__
print(repr(p)) # calls __repr__
```
