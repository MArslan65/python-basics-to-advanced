# Generators

```python
# generator function
def my_generator():
    yield 'A'
    yield 'B'
    yield 'C'
    yield 'D'
# create a generator object
gen = my_generator()
# iterate through the generator
for value in gen:
    print(value)
```

```python
# generator expression
squares = (x**2 for x in range(10))
# iterate through the generator expression
for square in squares:
    print(square)
```
