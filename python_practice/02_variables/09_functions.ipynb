## User defined functions

```python
def greet(name):
    """this function greets the person whose name is passed as an argument"""
    print("Hello," + name + ". How are you?")
```

```python
greet("Alice")
```

```python
def calculate_mean_price(prices):
    """this function calculates the mean price from a list of prices"""
    mean_price = round(sum(prices) / len(prices), 2)
    return mean_price
```

```python
prices =  [200, 300, 400, 500, 600,400,5,500,600,400,500,600]
mean_price = calculate_mean_price(prices)
mean_price
```

```python
?mean_price
```

```python
def add(a, b):
    """this function adds two numbers and returns the result"""
    return a + b
```

```python
add(5, 10)
```

```python
def introduce_yourself(name, age):
    """this function introduces yourself by printing your name and age"""
    print("Hello, my name is " + name + " and I am " + str(age) + " years old.")
```

```python
introduce_yourself("Alice", 30)
```

# functions without parameters

```python
def say_hello():
    """this function prints hello"""
    print("Hello")
```

```python
say_hello()
```

```python
# function with default parameter value
def greet(name="stranger"):
    """this function greets the person whose name is passed as an argument, if no name is passed, it greets a stranger"""
    print("Hello," + name + ". How are you?")
```

```python
greet("bob")
```
