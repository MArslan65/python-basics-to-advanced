 # IF else


```python
AGE = 10
if AGE >= 18:
    print("You are eligible to vote.")
elif AGE == 17 :
    print("You will be eligible to vote next year.")
elif AGE == 16 and AGE < 18:
    print("You will be eligible to vote in two years.")
else:
    print("You are not eligible to vote.")
```

## nested if else

```python
age = 20
license = False

if age >= 18:
    license = True
    if license:
        print("You can drive.") 
    else:
        print("You need a license to drive.")
else:
    print("You are not old enough to drive.")
```

# while loop

```python
cup = 0
while cup < 10:
    print("The cup is not full yet.")
    cup += 1
print("The cup is full.")
```

# break and continue

```python
cup = 0
while cup < 5:
    cup += 1
    print("The cup is not full yet.")
    if cup == 3:
        print("The cup is full.")
        break
    print ("enough", cup, "cup.")   
```

# continue

```python
cup = 0
while cup < 5:
    cup += 1
    print("The cup is not full yet.")
    if cup == 3:
        print("The cup is full.")
        continue
    print ("enough", cup, "cup.")   
```

# range function in loops

```python
for cup in range(1,6,2):
    print("chai is ready",cup, "served.")
```

# infinate loop

```python
while True:
    print("This loop will run forever.")
    break
```

```python
counter = 0
while True:
    print("This loop will run forever.")
    counter += 1
    if counter >= 5:
        break
```

# pass statement

```python
cup = 0
while cup < 5:
    cup += 1
    print("The cup is not full yet.")
    if cup == 3:
        print("The cup is full.")
        pass
    print ("enough", cup, "cup.")   
```

```python
# example of try and except finally
try:
    print("Hello")
    print(1/0)  # This will raise a ZeroDivisionError
except ZeroDivisionError:
    print("You cannot divide by zero.")
finally:
    print("This will always execute.")
```
