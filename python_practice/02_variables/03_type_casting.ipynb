# Type Casting

## Int to Float

```python
age_int= 29
age_float = float(age_int)
print(type(age_float))
print(age_float)
```

## Float to Int

```python
age_float = 29.53
age_int = int(age_float)
print(type(age_int))
print(age_int)
```

## float or int to String

```python
print("INT to FLOAT")
age_int= 29
age_float =29.5
```

```python
age_int = str(age_int)
age_float = str(age_float)
print(type(age_int))
```

## string to int and float

```python
value = "29"
value = int(value)
value_float = float(value)
```

## List to tuple and tuple to list

```python
my_list = [1, 2, 3, 4, 5]
my_tuple = (1, 2, 3, 4, 5)
print(type(my_list))
print(type(my_tuple))
```

```python
my_list_to_tuple = tuple(my_list)
print(type(my_list_to_tuple))   
my_tuple_to_list = list(my_tuple)
print(type(my_tuple_to_list))
```

## list to set and set to list

```python
myh_list = [1, 2, 3, 4, 5]
myh_list_set = set(myh_list)
print(type(myh_list_set))
```

## to boolean
