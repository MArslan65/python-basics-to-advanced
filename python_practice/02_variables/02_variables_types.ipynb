# types of variable

# 1. Integer

```python
age = 10
print(type(age))
```

# 2. Floating

```python
temperature = 36.5
print(type(temperature))
```

# 3. string

```python
greeting = "Hello, World!"
print(type(greeting))
```

# 4. Boolean

```python
is_raining = True
type(is_raining)
```

# 5.list

### Key Features
✔ Ordered → items keep their position
✔ Mutable → you can change it
✔ Allows duplicates

```python
number = [1, 2, 3, 4, 5]
print(type(number))
person =["name", "age", "city"]
print(type(person))
```

# 6. tuple

### key feature
✔ Ordered
❌ Not mutable (cannot change)
✔ Allows duplicates

```python
price = (19.99, 29.99, 39.99)
print(type(price))
```

# 7. dictionary

### key feature
✅ Key Features
✔ Key-Value structure
✔ Mutable (can change)
✔ Keys must be unique
✔ Values can be anything

```python
person_info = {"name": "Alice", "age": 30, "city": "New York"}
print(type(person_info))
print(person_info)
```

# 8. set

### key feature
✅ Key Features
❌ Unordered (no index)
✔ Mutable (can add/remove)
❌ No duplicates

```python
fruits = {"apple", "banana", "orange"}
print(type(fruits))
print(fruits)
```

# 9. NoneType

### key feature
Where used?
Function returns nothing
Placeholder variable
Checking missing values

```python
result = None
print(type(result))
```

# 10. Complex

### Complex → represents numbers with real + imaginary part (a + bj)

```python
complex_number = 2 + 3j
print(type(complex_number))
```

# Quick comparison between list,tuple,set
| Feature    | List | Tuple | Set |
| ---------- | ---- | ----- | --- |
| Mutable    | ✅    | ❌     | ✅   |
| Duplicates | ✅    | ✅     | ❌   |
| Indexing   | ✅    | ✅     | ❌   |
| Order      | ✅    | ✅     | ❌   |
